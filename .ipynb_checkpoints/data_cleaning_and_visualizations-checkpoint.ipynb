```python
# Chicago Socioeconomic Data Analysis

# ===============================
# STEP 1: Import Required Libraries
# ===============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3
import os
```

```python
# Create directories if not exist
os.makedirs("data", exist_ok=True)
os.makedirs("database", exist_ok=True)
os.makedirs("visualizations", exist_ok=True)
```

```python
path ="./data/chicago_data_raw.csv"
df = pd.read_csv(path)
df
```

```python
# Data Cleaning
# Clean column names
df.columns = df.columns.str.lower().str.replace(' ', '_')

# Replace inf with NaN and drop missing rows
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)
df
```

```python
# Save cleaned data to 'data/' folder
df.to_csv("data/chicago_data_cleaned.csv", index=False)  
```

```python
# Create SQLite database in 'database/' folder
import csv, sqlite3
conn = sqlite3.connect("database/chicago.db")  # ðŸ”„ database/
df.to_sql("chicago_data", conn, if_exists="replace", index=False,method="multi")
print("\nTable 'chicago_data' saved to database.")  # Stored inside chicago.db
```

```python
# Run SQL queries on the database
print("\n--- SQL: Top 5 Poorest Areas ---")
query1 = """
SELECT community_area_name, percent_households_below_poverty
FROM chicago_data
ORDER BY percent_households_below_poverty DESC
LIMIT 5;
"""
print(pd.read_sql_query(query1, conn))  # From database/
```

```python
print("\n--- SQL: Top 5 Highest Income Areas ---")
query2 = """
SELECT community_area_name, per_capita_income_
FROM chicago_data
ORDER BY per_capita_income_ DESC
LIMIT 5;
"""
print(pd.read_sql_query(query2, conn))  # From database/
```

```python
os.makedirs("visualizations1", exist_ok=True)
```

```python
# Set Seaborn style
sns.set(style="whitegrid")

# Scatterplot: Income vs Poverty
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='per_capita_income_', y='percent_households_below_poverty')
plt.title("Income vs. Poverty in Chicago Community Areas")
plt.xlabel("Per Capita Income")
plt.ylabel("% Households Below Poverty")
plt.savefig("visualizations1/income_vs_poverty.png")  # ðŸ”„ visualizations/
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Socioeconomic Factors")
plt.savefig("visualizations1/correlation_matrix.png")  # ðŸ”„ visualizations/
plt.show()

```

```python
# Barplot: Top 10 Areas with Highest Unemployment
top_unemployed = df.nlargest(10, 'percent_aged_16+_unemployed')
plt.figure(figsize=(10, 6))
sns.barplot(data=top_unemployed, x='percent_aged_16+_unemployed', y='community_area_name', palette="Reds_r")
plt.xlabel("% Unemployed Age 16+")
plt.ylabel("Community Area")
plt.title("Top 10 Areas with Highest Unemployment")
plt.tight_layout()
plt.savefig("visualizations1/top_unemployment.png")  # ðŸ”„ visualizations/
plt.show()
```
